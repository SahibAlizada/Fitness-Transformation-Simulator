```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\Sahib Alizada\Desktop\SHB\Real Projects\7-Fitness Progress & Body Transformation Analytics\Data\gym_members_exercise_tracking.csv")
```

```python
df["BMI_Calc"] = df["Weight (kg)"] / (df["Height (m)"] ** 2)

(df["BMI"] - df["BMI_Calc"]).abs().mean()
```

```python
df = df[(df["BMI"] > 15) & (df["BMI"] < 45)]
```

```python
df = df[df["Fat_Percentage"] > 15]
```

```python
df["Weekly_Calories"] = df["Calories_Burned"] * df["Workout_Frequency (days/week)"]
```

```python
df["Monthly_Calories"] = df["Weekly_Calories"] * 4
```

```python
df["Monthly_Fat_Loss_kg"] = df["Monthly_Calories"] / 7700
```

```python
df["Current_Fat_kg"] = df["Weight (kg)"] * (df["Fat_Percentage"] / 100)
```

```python
df["Target_Fat_kg"] = df["Weight (kg)"] * 0.15
```

```python
df["Fat_kg_To_Lose"] = df["Current_Fat_kg"] - df["Target_Fat_kg"]
```

```python
df["Months_To_Target"] = df["Fat_kg_To_Lose"] / df["Monthly_Fat_Loss_kg"]
```

```python
df = df[(df["Months_To_Target"] > 0) & (df["Months_To_Target"] < 24)]
```

```python
df["Months_To_Target"].mean()
```

```python
df.groupby("Workout_Frequency (days/week)")["Months_To_Target"].mean()
```

```python
df[df["Workout_Frequency (days/week)"] == 5][["Fat_Percentage"]].describe()
```

```python
df_compare = df[df["Fat_Percentage"] > 20]

df_compare.groupby("Workout_Frequency (days/week)")["Months_To_Target"].mean()
```
